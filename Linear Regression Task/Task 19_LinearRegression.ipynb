```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#import our data
data =pd.read_csv('insurance.csv')
data.head()
```

```python
#get an overview of our data
data.describe()
```

```python
#understanding our dataset
data.info
```

```python
#plot a scatter plot with age on the x-axis and charges on the y-axis
x = data.iloc[:, 0].values
y = data.iloc[:, 6].values

plt.scatter(x,y, color= 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#Using linear_model.LinearRegression() from sklearn, we are fitting a model to our data and making predictions

# Create the linear regression model
model_data = LinearRegression()


# Reshape x from 1D to 2D
x = x.reshape(-1, 1)  
model_data.fit(x, y)




# Predict using the fitted model
y_pred = model_data.predict(x)



#plot predicted scatter plot
plt.scatter(x,y, color= 'r')
plt.plot(x,y_pred,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#Prediction on data 
print(f" Our predicted values are {y_pred}")
```

```python
#Plot another scatter plot with the best fit line

# Plot the scatter plot
plt.scatter(x, y, label='Actual Data', color = 'r')
plt.plot(x, y_pred, color='b', label='Best-Fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot with Best-Fit Line: Age vs Charges')
plt.legend()
plt.show()

```
